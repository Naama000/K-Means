```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs #makes artificials datasets in blobs like form
```

```python
data = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.8, random_state=101)
```

```python
data #thisis not a dataframe, yet. it's a tuple.
```

```python
data[0].shape # data[0] is where the features are stored. data[1] are the classes
```

```python
plt.scatter(data[0][:, 0], data[0][:,1], c = data[1], cmap = 'plasma') #data[1] are the classes and by which we're coloring the points
```

```python
data[1]
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])  # we're fitting the model with the features
```

```python
kmeans.cluster_centers_ #the predicted centeroids
```

```python
kmeans.labels_ #the labels the algorithem thinks
```

we can see there's a difference between the original and the predicted labels

```python
fig, (ax1, ax2) = plt.subplots(1,2, sharey = True, figsize = (10,6))
ax1.set_title('K Means')
ax1.scatter(data[0][:,0], data[0][:,1], c = kmeans.labels_, cmap = 'magma')
ax2.set_title('Original')
ax2.scatter(data[0][:,0], data[0][:,1], c = data[1], cmap = 'magma')
```

```python
kmeans2 = KMeans(n_clusters=2)
```

```python
kmeans2.fit(data[0])  # we're fitting the model with the features
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, sharey = True, figsize = (10,6))
ax1.set_title('K Means 2 clusters')
ax1.scatter(data[0][:,0], data[0][:,1], c = kmeans2.labels_, cmap = 'magma')
ax2.set_title('Original')
ax2.scatter(data[0][:,0], data[0][:,1], c = data[1], cmap = 'magma')
```
